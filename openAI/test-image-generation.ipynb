```python
import os
import openai
response = openai.Image.create(
  prompt="a teenage girl of afghani descent with striking blue eyes (as in dune) stares at the camera with a deep read head scarf. kodachrome film",
  n=1,
  size="1024x1024"
)
image_url = response['data'][0]['url']
#download the image
import requests
from PIL import Image
from io import BytesIO
response = requests.get(image_url)
img = Image.open(BytesIO(response.content))
img
#Save the image
img.save("test2.jpg")
```
